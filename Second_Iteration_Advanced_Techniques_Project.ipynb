```python
#Validation

import os
import torch
from torchvision import transforms
import PIL
PIL.ImageFile.LOAD_TRUNCATED_IMAGES = True
force_reload=True

# Load the trained model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/yolov5/runs/train/exp/weights/best.pt')

# Set the model to evaluation mode
model.eval()

# Directory containing the images
image_dir = '/content/ComputerPeripherals/valid/images/'

# Directory containing the labels
label_dir = '/content/ComputerPeripherals/valid/labels/'

# Get a list of all image files
image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]

# Iterate over each image file
for image_file in image_files:
    # Split the file name and extension
    base_name, extension = os.path.splitext(image_file)

    # If there's additional information after the extension, remove it
    if '.' in base_name:
        base_name = base_name.split('.')[0]

    # New file name with the corrected base name and extension
    new_image_file = base_name + extension

    # Rename the image file
    os.rename(os.path.join(image_dir, image_file), os.path.join(image_dir, new_image_file))

    # Now, find and rename the corresponding label file
    label_file = base_name + '.txt'  # Assuming the label files are .txt files
    new_label_file = new_image_file.split('.')[0] + '.txt'  # New label file name

    # Check if the label file exists
    if os.path.exists(os.path.join(label_dir, label_file)):
        # Rename the label file
        os.rename(os.path.join(label_dir, label_file), os.path.join(label_dir, new_label_file))
    else:
        print(f"Label file not found for image: {image_file}")

# Path to the directory containing YOLO format annotation files
#ann_dir = '/content/ComputerPeripherals/valid/labels/'

# List of image files
img_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif'))]

# Evaluate the model
predictions = []
for img_file in img_files:
    img_path = os.path.join(image_dir, img_file)
    print(img_path)
    label_path = os.path.join(label_dir, img_file.replace(os.path.splitext(img_file)[1], '.txt'))

    # Check if label file exists
    if os.path.exists(label_path):
        try:
            # Load image
            image = PIL.Image.open(img_path).convert("RGB")

            # Transform image
            transform = transforms.Compose([
                transforms.Resize((640, 640)),  # Model expects input images of size 640x640
                transforms.ToTensor()
            ])
            image_tensor = transform(image).unsqueeze(0)

            # Forward pass
            with torch.no_grad():
                outputs = model(image_tensor)

            # Convert predictions to YOLO format
            for output in outputs:
                for detection in output:
                    if detection is not None and len(detection) > 0:
                        for det in detection:
                            if det.numel() >= 4:  # Check if det has at least 4 elements
                                # Convert box format from (x1, y1, x2, y2) to (x_center, y_center, width, height)
                                bbox = det[:4].tolist()
                                bbox[0] = (bbox[0] + bbox[2]) / 2.0
                                bbox[1] = (bbox[1] + bbox[3]) / 2.0
                                bbox[2] = bbox[2] - bbox[0]
                                bbox[3] = bbox[3] - bbox[1]
                                predictions.append([0] + bbox)  # Assuming class index is 0
                            else:
                                print("Detection tensor does not contain enough elements.")
                    else:
                        print("No objects detected in image.")
        except PIL.UnidentifiedImageError:
            print(f"Unable to open image file: {img_path}")
    else:
        print(f"No label file found for image: {img_file}")

# Write predictions to a text file in YOLO format
with open('predictions.txt', 'w') as f:
    for pred in predictions:
        f.write(' '.join(map(str, pred)) + '\n')

# Ideally I should now have a usuable predictions.txt file to use later

```

```python
#Run the model on a video - currently it is not working at the validation step, which I tried AFTER originally attempting to go straight from training to use. This was not smart and I am now stuck at validation)

# Import necessary libraries
import cv2
import torch

# Load the model from the Hub
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/yolov5/runs/train/exp2/weights/best.pt')

# Set the model to evaluation mode
model.eval()

# Move the model to GPU if available
if torch.cuda.is_available():
    model.to('cuda')

# Open the video file
cap = cv2.VideoCapture('/content/PokimaneSetupTourCompressed.mp4')

# Check if the video file is opened correctly
if not cap.isOpened():
    print("Error opening video file")
    exit()

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Check if the frame is not empty
    if not ret:
        print("End of video or error capturing frame")
        break

    # Convert the frame to RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Convert the frame to a tensor
    frame_tensor = torch.from_numpy(frame_rgb).float() / 255.0

    # Add a batch dimension
    frame_tensor = frame_tensor.unsqueeze(0)

    # Move the tensor to GPU if available
    if torch.cuda.is_available():
        frame_tensor = frame_tensor.to('cuda')

    # Perform object detection
    with torch.no_grad():
        predictions = model(frame_tensor)

    # Filter predictions for computer peripherals
    peripherals = ['headset', 'keyboard', 'mouse']
    predictions = [pred for pred in predictions if pred['label'] in peripherals]

    # Draw bounding boxes and labels on the frame
    for pred in predictions:
        x1, y1, x2, y2 = pred['box']
        label = pred['label']
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Convert the frame back to BGR for display
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

    # Display the resulting frame
    cv2.imshow('frame', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When finished, release the capture
cap.release()
cv2.destroyAllWindows()
```
